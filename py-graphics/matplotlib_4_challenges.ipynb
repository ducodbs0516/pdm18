> [hint] 검색: "How to add title to subplots in Matplotlib?"

```
import numpy as np
import matplotlib.pyplot as plt
```

```
plt.subplot(2,1,1)
plt.hist(np.random.rand(1000))
plt.title(" rand (uniform distribution)")
plt.xlabel("x")
plt.ylabel("y")
plt.subplots(constrained_layout=True) # 서브플롯 간의 간격을 가능한 최적의 수치로 자동 조정

plt.subplot(212)
plt.hist(np.random.randn(1000))
plt.title(" randn (normal distribution, Gaussian)")
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```

```
fig, ax = plt.subplots(1,2, figsize=(16, 8))

ax[0].hist(np.random.rand(1000))
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[0].set_title('rand (uniform distribution)')

ax[1].hist(np.random.randn(1000))
ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')
ax[1].set_title('randn (normal distribution, Gaussian)')

plt.show()
```

- uniform distribution subplots

```
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=False, figsize=(10, 10))
axs = [ax1,ax2,ax3,ax4]

for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.rand(sample_size)
    axs[n].hist(sample)
    axs[n].set_xlabel('X')
    axs[n].set_ylabel('Y')
    axs[n].set_title('n={}'.format(sample_size))
```

- normal distribution subplots

```
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=False, figsize=(10, 10))
axs = [ax1,ax2,ax3,ax4]

for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.randn(sample_size)
    axs[n].hist(sample)
    axs[n].set_xlabel('X')
    axs[n].set_ylabel('Y')
    axs[n].set_title('n={}'.format(sample_size))
```

### [도전 2.] 0~9 사이의 무작위 정수 1000개를 만들어서 각 수의 빈도에 대한 히스토그램, 막대그래프, 꺽은선 그래프를 그리고 비교하시오.
- subplot 이용
- style(제목, x-축 레이블, y-축 레이블, ...)



```
n1000 = np.random.randint(0,10,1000)
```

```
value, count= np.unique(n1000, return_counts=True)
value, count
```

```
import numpy as np
import matplotlib.pyplot as plt
```

```
plt.subplot(411)
plt.hist(count,bins=20)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Histogram')
plt.subplots(constrained_layout=True) # 서브플롯 간의 간격을 가능한 최적의 수치로 자동 조정

plt.subplot(412)
plt.bar(value,count) #,color="bgrcmyk")  # 'bgrcmykw
plt.xlabel('X')
plt.ylabel('Y')
plt.title('bar graph')
plt.subplots(constrained_layout=True) 

plt.subplot(413)
plt.plot(value,count)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('line graph')
plt.subplots(constrained_layout=True)

plt.subplot(414)
plt.plot(value,count,'-o')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('line graph')
plt.show()
```

- subplots

```
fig, ax = plt.subplots(1,3, figsize=(24, 8))

ax[0].hist(count,bins=20)
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')
ax[0].set_title('Histogram')

ax[1].bar(value,count)
ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')
ax[1].set_title('bar graph')

ax[2].plot(value,count,'-o')
ax[2].set_xlabel('X')
ax[2].set_ylabel('Y')
ax[2].set_title('line graph')

plt.show()
```

- Histogram subplots

```
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, sharex=False, figsize=(10,5))
axs = [ax1,ax2,ax3,ax4]

for n in range(0,len(axs)):
    sample_size = 10**(n+1)
    sample = np.random.randint(0,10,1000)
    axs[n].hist(sample,bins=30)
    axs[n].set_xlabel('X')
    axs[n].set_ylabel('Y')
    axs[n].set_title('n={}'.format(sample_size))
```
